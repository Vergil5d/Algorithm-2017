```python
import matplotlib.pyplot as plt
import numpy as np

FD=100.0
FS = 1/FD
x = np.arange(0,10,FS)
plt.plot(x)
plt.ylabel('initial data')
plt.show()

def fft(x):
    if (len(x) == 1):
        return x
    even = x[0::2]
    odd = x[1::2]
    a = fft(even)
    b = fft(odd)
    N = len(a) * 2
    w = np.exp(-2j * np.pi / N * np.arange(len(a)))
    r0 = a + b * w
    r1 = a - b * w
    r = np.append(r0, r1)
    return r

def ifft(x):
    val = fft([xi.conjugate() for xi in x])
    return [xi.conjugate()/len(val) for xi in val]

g=fft(x)

D=100

plt.plot(g[:D],'g')
plt.ylabel('FFT from X')
plt.show()
g2= ifft(g)

plt.plot(g2[:D],'g')
plt.ylabel('IFFT from X')
plt.show()

```

```python
FD=100.0
FS = 1/FD
x = np.arange(0,10,FS)

value = [xi**2 for xi in x]

plt.plot(value[:D])
plt.ylabel('initial data')
plt.show()

g=fft(value)

D =100
plt.plot(g[:D],'g')
plt.ylabel('FFT from x^2 ')
plt.show()

g2= ifft(g)

plt.plot(g2[:D],'g')
plt.ylabel('IFFT from x^2')
plt.show()
```
